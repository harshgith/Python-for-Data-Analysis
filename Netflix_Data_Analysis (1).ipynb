```python
import numpy as np # linear algebra
import pandas as pd # data processing
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
netflix=pd.read_csv("https://raw.githubusercontent.com/sachinkumar1609/Netflix-Data-Analysis/main/netflix_titles.csv")
netflix.head(10)
```

```python
netflix.duplicated().sum()#check if any duplicate values are present in the Database
```

```python
df=netflix.copy()#make a copy of the dataset
df.shape
```

```python
df=df.dropna()#drop null values from dataset
df.shape
```

```python
df.head(10)#Much bettr and Cleaned Dataset
```

```python
#convert the date and time format

df["date_added"] = pd.to_datetime(df['date_added'])
df['day_added'] = df['date_added'].dt.day
df['year_added'] = df['date_added'].dt.year
df['month_added']=df['date_added'].dt.month
df['year_added'].astype(int);
df['day_added'].astype(int);
```

```python
df.head(10)
```

```python
sns.countplot(netflix['type'])
fig = plt.gcf()
fig.set_size_inches(10,10)
plt.title('Type')
```

```python
sns.countplot(netflix['release_year'])
fig = plt.gcf()
fig.set_size_inches(40,30)
plt.title('release_year')
```

From the Above Analysis We can observe that Most of the Films and Tv Shows were released in the year 2019 with an count of over 1000

```python
#Rating of shows and movies

sns.countplot(netflix['rating'])
sns.countplot(netflix['rating']).set_xticklabels(sns.countplot(netflix['rating']).get_xticklabels(), rotation=90, ha="right")
fig = plt.gcf()
fig.set_size_inches(13,13)
plt.title('Rating')
```

```python
#Relation between Type and Rating
plt.figure(figsize=(20,10))
sns.countplot(x='rating',hue='type',data=netflix)
plt.title('Relation between Type and Rating')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.countplot(x='rating',hue='type',data=netflix)
plt.title('Relation between Type and Rating')
plt.show()
```

```python
#Pie-chart for the Type: Movie and TV Shows

labels = ['Movie', 'TV show']
size = netflix['type'].value_counts()
colors = plt.cm.Wistia(np.linspace(0, 1, 2))
explode = [0, 0.1]
plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size,labels=labels, colors = colors, explode = explode, shadow = True, startangle = 90)
plt.title('Distribution of Type', fontsize = 25)
plt.legend()
plt.show()
```

Now,Let us Simplify Further the pie chart and analysis, we can see from the image below that 68.4% of dataset are of Movies Type and rest 31.6% are of Tv Shows Type

```python
import plotly.graph_objs as go

label = ['Movie', 'TV Show']
counts = netflix['type'].value_counts(normalize=True)
colors = ['gold', 'lightgreen']

fig = go.Figure(data=[go.Pie(labels=label, values=counts, marker=dict(colors=colors, line=dict(color='black', width=3)))])
fig.update_layout(title_text='Proportion of Movies and TV Shows on Netflix')
fig.show()
```

```python
#Pie-chart for Rating
netflix['rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True,figsize=(15,10))
plt.show()
```

```python
plt.subplots(figsize=(20,15))
wordcloud = WordCloud(
                          background_color='Black',
                          width=1920,
                          height=1080
                         ).generate(" ".join(df.country))
plt.imshow(wordcloud)
plt.axis('off')
plt.savefig('country.png')
plt.show()
```
